```python
#import method from library
from sklearn.model_selection import train_test_split
```

```python
#independent variable
x=maharashtra['death']
```

```python
#we cant normally use date we have to convert it in date time...it should be in sequence numbers
```

```python
#dependent variable
y=maharashtra['confirm']
```

```python
#split data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#import linear regression
from sklearn.linear_model import LinearRegression
```

```python
#create instance
lr=LinearRegression()
```

```python
y_train
```

```python
#here we can see dataframes are labelled to avoid this we have to reshape while fitting
```

```python
#fit the model on training dataset
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
y_predict=lr.predict(np.array(x_test).reshape(-1,1))
```

```python
y_test.head()
```

```python
y_predict[0:5]
```

```python
#to get mean squared error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_predict)
```
